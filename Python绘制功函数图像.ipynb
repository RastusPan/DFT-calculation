```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 文件路径
file_path = input("Input the address of the Vline file:")
```

```python
point = input("The name of the high-symmetric point:")
```

```python
z_len = float(input("The len of the z-axis:"))
```

```python
# 读取文件
with open(file_path, 'r') as file:
    # 读取第一行
    first_line = file.readline()
    
    # 分割第一行的数据
    parts = first_line.split()
    
    # 提取第一行的第一个数据
    first_data = parts[0]

    # 输出第一行的第一个数据
    print("第一行的第一个数据:", first_data)

    # 跳过第一行
    next(file)
    
    # 初始化空列表来存储数据
    keys = []
    values = []

    # 逐行读取数据
    for line in file:
        # 分割每行的数据
        parts = line.split()
        
        # 提取key和value
        key = parts[0]
        value = float(parts[1])  # 假设value是浮点数

        # 将数据添加到列表中
        keys.append(key)
        values.append(value)
```

```python
smooth_keys = np.linspace(0, len(keys)-1, 1000)  # 生成1000个均匀分布的点
smooth_values = np.interp(smooth_keys, range(len(keys)), values)
```

```python
x_values = z_len * smooth_keys / int(first_data)
```

```python
y_coord = float(input("Fermi Energy:"))
label_text = "E-fermi = " + str(y_coord)
```

```python
# 绘图
plt.plot(x_values, smooth_values)
plt.xlabel('Position (A)')
plt.ylabel('Energy (eV)')
plt.title(point)
plt.ylim(-25, 10)
plt.xlim(0, 20)
#zhushi ='x = %s,y = %s'%(len(keys) / 2,round(y_coord - 0.1,3)) 
plt.axhline(y=y_coord, color='red', linestyle='--')
# 在虚线下方添加标注文字
plt.text(15, y_coord - 1.4, label_text, color='red', fontsize=12, ha='center', va='center')
#plt.legend()

plt.show()
```
